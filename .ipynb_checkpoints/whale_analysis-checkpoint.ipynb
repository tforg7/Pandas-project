# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_path = Path("./Resources/whale_returns.csv")
df_whale_return = pd.read_csv(whale_path, index_col="Date", infer_datetime_format =True, parse_dates=True)
```

```python
# Count nulls : counts empty values and NA.
df_whale_return.isnull().sum()
```

```python
# Drop nulls: remoce the NA and empty values from the dataframe
df_whale_return.dropna(inplace=True)

# Proceed to a data check, columns need to be floats and null need to be dropped(entries need to egal non-null count)
df_whale_return.info()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_path = Path("./Resources/algo_returns.csv")
# Put the csv into a dataframe for analyse, put the date contented in the csv immediately as the index
df_algo_return = pd.read_csv(algo_path, index_col="Date", infer_datetime_format =True, parse_dates=True)
```

```python
# Count nulls : counts empty values and NA.
df_algo_return.isnull().sum()
```

```python
# Drop nulls: remoce the NA and empty values from the dataframe
df_algo_return.dropna(inplace=True)

# Proceed to a data check, columns need to be floats and null need to be dropped(entries need to egal non-null count). Start date and date is also intresting in order to have an idea of the timeframe
df_algo_return.info()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_path = Path("./Resources/sp500_history.csv")
# Put the csv into a dataframe for analyse, put the date contented in the csv immediately as the index
df_sp500 = pd.read_csv(sp500_path, index_col="Date", infer_datetime_format =True, parse_dates=True)
```

```python
# Check Data Types
df_sp500.dtypes
```

```python
# Fix Data Types : First removing the $ signe and then change the string into a float value
df_sp500["Close"] = df_sp500["Close"].str.replace("$", "")
df_sp500["Close"] = df_sp500["Close"].astype(float)

# Control aspect of the dataframe
df_sp500.head()
```

```python
# Calculate Daily Returns with the function pct_change()
df_sp500_return = df_sp500.pct_change()

# Visualize the dataframe df_sp500_return
df_sp500_return.head()
```

```python
# Drop nulls : first row has NA value after using the function pct_change()
df_sp500_return.dropna(inplace=True)

# Proceed to a data check, columns need to be floats and null need to be dropped(entries need to egal non-null count). Start date and date is also intresting in order to have an idea of the timeframe
df_sp500_return.info()
```

```python
# Rename `Close` Column to be specific to this portfolio.
df_sp500_return.rename(columns = {
                        "Close" : "S&P 500"
                        }, inplace=True)
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
return_df = pd.concat([df_whale_return, df_algo_return, df_sp500_return ], axis="columns" , join = "inner")
return_df.info()
```

### Create a box plot for each portfolio

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
std_return = return_df.std()
std_return
```

### Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500 (only over the timeframe of the dataframe return_df in order to compare over the identic timeframe for every portfolio)
daily_std_sp500 = std_return["S&P 500"]

# Determine which portfolios are riskier than the S&P 500 : portfolios with higher standard deviation than s&p 500 are riskier
std_return.sort_values(ascending=False, inplace=True)
risky_portfolio = std_return.index[std_return > daily_std_sp500 ].tolist()


# print portfolio's that are riskier than the S&P 500
print(f"The portfolio's {risky_portfolio} have a bigger standard deviation than the S&P 500 ({daily_std_sp500}), which make then riskier than the S&P 500")
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
volatility = std_return * np.sqrt(252)
volatility.sort_values(ascending=False, inplace=True)
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
std_return_21 = return_df.rolling(window=21).std().dropna()

# Plot the rolling standard deviation
std_return_21.plot(figsize= (20,7), title ="Rolling standard deviation using a 21-day window", ylabel = "standard deviation", ylim = (0,.035))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation_return = return_df.corr()
# Display de correlation matrix, how ever I considered that a heatmap is a better visualisation and displayed the result with a heatmap here under
correlation_return.head(7)

# Display of a heatmap in order to visualize easely the correlation with colours
sns.heatmap(correlation_return, vmin=-1, vmax=1)

# Determine which portfolio has the most corroletion with the others porfolios. The higher the absolute sum, the higher the corrolation in general. This is a rough approximation.
correlation_return.abs().sum().sort_values()
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
cov_algo1 = return_df["Algo 1"].cov(return_df["S&P 500"])
# Calculate variance of S&P 500
variance_sp500 = return_df["S&P 500"].var()
# Computing beta
beta_algo1 = cov_algo1 / variance_sp500

# Plot beta trend : using a rolling window of 21 days
cov_rolling_algo1 = return_df["Algo 1"].rolling(window=21).cov(return_df["S&P 500"])
variance_rolling_sp500 = return_df["S&P 500"].rolling(window=21).var()

beta_rolling_algo1 = cov_rolling_algo1 / variance_rolling_sp500
beta_rolling_algo1.plot(figsize= (10,5), title =" Rolling beta of portfolio Algo 1 with a rolling window of 21 days", ylabel ="Beta")
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
std_ewm_21 = return_df.ewm(halflife=21.0).std()

#comparing the two methodes on the return from the protfolio Algo 1 : rolling statistics and exponentially weigthed moving average with both 21 days.
ax = std_return_21["Algo 1"].plot(figsize= (20,10), color="red")
std_ewm_21["Algo 1"].plot(figsize= (20,10), title ="Rolling statistics Vs. Exponentially Weighted Average with a 21-day half-life", subplots = True, ylim = (0,.025), ax=ax)
ax.legend(["Rolling statistics", "Exponentially Weighted Average"])
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot






```python
# Annualized Sharpe Ratios
# Approximate the risk free rate with the csv used in an exercice of previous weeks.
risk_free_rate_path= Path("./Resources/risk_free_rate.csv")
risk_free_rate = pd.read_csv(risk_free_rate_path, index_col='Date', parse_dates=True, infer_datetime_format=True)

# Take the dates from the first and last index of the dataframe return_df in order to compare over the same periode
dates = return_df.index[[0, -1]].tolist()

risk_free_rate = risk_free_rate.loc[dates[0]: dates[1]]
risk_free_rate.head()
```

```python
# determine the sharp ratios. The annual standard deviation is alreadu calculated in the variable 'volatility'. Then sort the values in order to read the result easely and finally show the result for the 7 portfolios
sharp_ratios =  ( (return_df.mean() - risk_free_rate["rf_rate"].mean())*252 ) / volatility
sharp_ratios.sort_values(ascending=False, inplace=True)
sharp_ratios.head(7)
```

```python
# Visualize the sharpe ratios as a bar plot
sharp_ratios.plot(kind="bar", figsize=(10,5))
```

```python
# Reading data from 1st stock
first_path = Path("./resources my portfolio/Amazon_stock.csv")
amazon = pd.read_csv(first_path)

# Date is a fusion of short date and time, we are spliting the two
date_amazon = amazon["Date"].str.split(" ", expand=True)

# Concat the first dataframe with the separate time and date
amazon_df = pd.concat([amazon, date_amazon[0]], axis="columns" , join="inner")
# Rename columns
amazon_df.columns = ["long_date" ,"AMZN" , "Date"]

# Modify short date to date format
amazon_df["Date"] =pd.to_datetime(amazon_df["Date"])

# Set short date as a new index and drop unnecessary columns
amazon_df.set_index(amazon_df["Date"], inplace=True)
amazon_df.drop(columns = ["long_date" ,"Date"], inplace=True)

# Fix Data Types : First removing the , signe and replace it by . sign then change the string into a float value
amazon_df["AMZN"] = amazon_df["AMZN"].str.replace(",", ".")
amazon_df["AMZN"] = amazon_df["AMZN"].astype(float)

amazon_df.head()               
```

```python
# Reading data from 2nd stock
second_path =Path("./resources my portfolio/IBM_stock.csv")
ibm = pd.read_csv(second_path)

# Date is a fusion of short date and time, we are spliting the two
date_ibm = ibm["Date"].str.split(" ", expand=True)

# Concat the first dataframe with the separate time and date
ibm_df = pd.concat([ibm, date_ibm[0]], axis="columns" , join="inner")
# Rename columns
ibm_df.columns = ["long_date" ,"IBM" , "Date"]

# Modify short date to date format
ibm_df["Date"] =pd.to_datetime(ibm_df["Date"])

# Set short date as a new index and drop unnecessary columns
ibm_df.set_index(ibm_df['Date'], inplace=True)
ibm_df.drop(columns = ["long_date" ,"Date"], inplace=True)

# Fix Data Types : First removing the , signe and replace it by . sign then change the string into a float value
ibm_df["IBM"] = ibm_df["IBM"].str.replace(",", ".")
ibm_df["IBM"] = ibm_df["IBM"].astype(float)

ibm_df.head()      
```
